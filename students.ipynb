```python
import pandas as pd
```

```python
# Load the CSV files into DataFrames
links = pd.read_csv('links.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
movie_ratings = pd.merge(ratings, movies, on='movieId', how='left')
```

```python
movie_ratings_tags = pd.merge(movie_ratings, tags, on=['userId', 'movieId'], how='left')
```

```python
final_df = pd.merge(movie_ratings_tags, links, on='movieId', how='left')
```

```python
final_df.dropna(axis=1, inplace=True)
final_df.head()
```

```python
# Transforms cleaned & updated csv file to excel

final_df = final_df.applymap(lambda x: x.encode('unicode_escape').decode('utf-8') if isinstance(x, str) else x)
```

```python
final_df.to_excel('output.xlsx', index=False)
```
